```python
class Estados(KnowledgeEngine):
    #@DefFacts()
    #def _initial_action(self):
        #yield Fact(action="start")

    
    #inicia o sistema
    @Rule(Fact(action='start'),
          NOT(Fact(ask=W())))
    def ask_start(self):
        self.declare(Fact(ask=input("Olá! Vamos identificar um estado do Brasil (S/N)?")))
    
    #inicia o sistema
    @Rule(Fact(action='restart'),
          NOT(Fact(ask=W())))
    def ask_restart(self):
        self.declare(Fact(ask=input("Deseja identificar outro Estado (S/N)?")))
    
    #sai do sistema
    @Rule(Fact(ask='N'))
    def tchau(self):
        print("Tchau.")
    
    
    #pergunta o clima do estado
    @Rule(OR(Fact(ask='S'),Fact(ask='s')),
          NOT(Fact(clima=W())))
    def ask_clima(self):
        self.declare(Fact(clima=input("Qual o clima do estado (subtropical, tropical, equatorial, semiárido)? ")))
    
    #pergunta a vegetação do estado
    @Rule(Fact(clima=W()),
          NOT(Fact(vegetacao=W())))
    def ask_vegetacao(self):
          self.declare(Fact(vegetacao=input("Qual a vegetação predominante do estado (caatinga, cerrado, pampa, pantanal, floresta amazônica, mata atlântica, mata subtropical)? ")))
            
    #mato grosso do sul
    @Rule(AND(Fact(clima='tropical'),
              Fact(vegetacao='pantanal')))
    def is_matogrossodosul(self):
          print('O estado é o Mato Grosso do Sul.')
            
    #rio grande do sul
    @Rule(AND(Fact(clima='subtropical'),
              Fact(vegetacao='pampa')))
    def is_riograndedosul(self):
          print('O estado é o Rio Grande do Sul.')
    
    #parana
    @Rule(AND(Fact(clima='subtropical'),
              Fact(vegetacao='mata subtropical')))
    def is_parana(self):
          print('O estado é o Paraná.')
            
    #Piauí
    @Rule(AND(Fact(clima='semiárido'),
              Fact(vegetacao='cerrado')))
    def is_piaui(self):
          print('O estado é o Piauí.')
            
    #pergunta se o estado tem mais de 40 milhoes de habitantes
    @Rule(AND(Fact(clima='subtropical'),
              Fact(vegetacao='mata atlântica')),
              NOT(Fact(subt_matl_habitantes_maisde40m=W())))
    def ask_hab_subtropical_mataatlantica(self):
        self.declare(Fact(subt_matl_habitantes_maisde40m=input("A população do estado tem mais de 40 milhões de habitantes (S/N)?")))
    
    #sao paulo
    @Rule(Fact(subt_matl_habitantes_maisde40m='S'))
    def is_saopaulo(self):
          print('O estado é São Paulo.')
    
    #santa catarina
    @Rule(Fact(subt_matl_habitantes_maisde40m='N'))
    def is_santacatarina(self):
          print('O estado é Santa Catarina.')
    
    #pergunta se o estado tem menos de 1 milhao de habitantes
    @Rule(AND(Fact(clima='equatorial'),
              Fact(vegetacao='floresta amazônica')),
              NOT(Fact(equa_famz_habitantes_menosde1m=W())))
    def ask_hab_equatorial_florestaamazonica(self):
        self.declare(Fact(equa_famz_habitantes_menosde1m=input("A população do estado tem menos de 1 milhão de habitantes (S/N)?")))
    
    #pergunta se o estado é banhado pelo oceano atlantico
    @Rule(Fact(equa_famz_habitantes_menosde1m=W()),
              NOT(Fact(oceanoatlantico=W())))
    def ask_oce_equatorial_florestaamazonica_menos1mhab(self):
        self.declare(Fact(oceanoatlantico=input("O estado é banhado pelo oceano atlantico? (S/N)?")))
    
    #amapa
    @Rule(AND(Fact(equa_famz_habitantes_menosde1m='S'),
          Fact(oceanoatlantico='S')))
    def is_amapa(self):
        print('O estado é o Amapá.')
    
    #para
    @Rule(AND(Fact(equa_famz_habitantes_menosde1m='N'),
          Fact(oceanoatlantico='S')))
    def is_para(self):
        print('O estado é o Pará.')
    
    #pergunta se o estado
    @Rule(AND(Fact(equa_famz_habitantes_menosde1m='S'),
              Fact(oceanoatlantico='N')),
              NOT(Fact(equa_famz_noce_area_maisde200km=W())))
    def ask_ter_equatorial_florestaamazonica_menos1mhab_naoocea(self):
        self.declare(Fact(equa_famz_noce_area_maisde200km=input("A área do estado tem mais de 200 mil km²? (S/N)?")))
    
    #Roraima
    @Rule(Fact(equa_famz_noce_area_maisde200km='S'))
    def is_rr(self):
        print('O estado é Roraima.')
   
    #acre
    @Rule(Fact(equa_famz_noce_area_maisde200km='N'))
    def is_acre(self):
        print('O estado é o Acre.')

    #pergunta se o estado
    @Rule(AND(Fact(equa_famz_habitantes_menosde1m='N'),
              Fact(oceanoatlantico='N')),
              NOT(Fact(equa_famz_noce_area_maisde1mkm=W())))
    def ask_ter_equatorial_florestaamazonica_mais1mhab_naoocea(self):
        self.declare(Fact(equa_famz_noce_area_maisde1mkm=input("A área do estado tem mais de 1 milhão km²? (S/N)?")))
    
    #Rondônia
    @Rule(Fact(equa_famz_noce_area_maisde1mkm='N'))
    def is_rondonia(self):
        print('O estado é Rondônia.')
    
    #amazonas
    @Rule(Fact(equa_famz_noce_area_maisde1mkm='S'))
    def is_amazonas(self):
        print('O estado é o Amazonas.')
        
    #pergunta se o estado é banhado pelo oceano
    @Rule(AND(Fact(clima='tropical'),
              Fact(vegetacao='floresta amazônica')),
              NOT(Fact(oceanoatlantico=W())))
    def ask_oce_tropical_florestaamazonica(self):
        self.declare(Fact(oceanoatlantico=input("O estado é banhado pelo oceano atlantico? (S/N)?")))
    
    #Maranhão
    @Rule(AND(Fact(clima='tropical'),
              Fact(vegetacao='floresta amazônica'),
              Fact(oceanoatlantico='S')))
    def is_maranhao(self):
          print('O estado é o Maranhão.')
    
    #Mato Grosso
    @Rule(AND(Fact(clima='tropical'),
              Fact(vegetacao='floresta amazônica'),
              Fact(oceanoatlantico='N')))
    def is_matogrosso(self):
          print('O estado é o Mato Grosso.')

    
    #pergunta se o estado tem mais de 10 milhoes de habitantes
    @Rule(AND(Fact(clima='tropical'),
              Fact(vegetacao='mata atlântica')),
              NOT(Fact(trop_matl_habitantes_maisde10m=W())))
    def ask_hab_tropical_mataatlantica(self):
        self.declare(Fact(trop_matl_habitantes_maisde10m=input("A população do estado tem mais de 10 milhões de habitantes (S/N)?")))
    
    #rio de janeiro
    @Rule(Fact(trop_matl_habitantes_maisde10m='S'))
    def is_riodejaneiro(self):
          print('O estado é o Rio de Janeiro.')
    
    #espirito santo
    @Rule(Fact(trop_matl_habitantes_maisde10m='N'))
    def is_espiritosanto(self):
          print('O estado é o Espírito Santo.')
            
    
    #pergunta se o estado tem menos de 7 milhoes de habitantes
    @Rule(AND(Fact(clima='tropical'),
              Fact(vegetacao='cerrado')),
              NOT(Fact(trop_cerr_habitantes_menosde7m=W())))
    def ask_hab_tropical_cerrado(self):
        self.declare(Fact(trop_cerr_habitantes_menosde7m=input("A população do estado tem menos de 7 milhões de habitantes (S/N)?")))
    
    #pergunta se a area do estado tem menos de 10 mil km²
    @Rule(Fact(trop_cerr_habitantes_menosde7m='S'),
          NOT(Fact(trop_cerr_menos7mhab_area_menos10km=W())))
    def ask_ter_tropical_cerrado_menos7mhab(self):
          self.declare(Fact(trop_cerr_menos7mhab_area_menos10km=input("A área do estado tem menos de 10 mil km²? (S/N)?")))
    
    #Distrito Federal
    @Rule(Fact(trop_cerr_menos7mhab_area_menos10km='S'))
    def is_distritofederal(self):
          print('O estado é o Distrito Federal.')
    
    #Tocantins
    @Rule(Fact(trop_cerr_menos7mhab_area_menos10km='N'))
    def is_tocantins(self):
          print('O estado é o Tocantins.')

    #pergunta se o estado tem mais de 500 municipios
    @Rule(Fact(trop_cerr_habitantes_menosde7m='N'),
          NOT(Fact(trop_cerr_mais7mhab_municipios_maisde500=W())))
    def ask_ter_tropical_cerrado_mais7mhab(self):
          self.declare(Fact(trop_cerr_mais7mhab_municipios_maisde500=input("O estado tem mais de 500 municipios? (S/N)?")))
    
        
    #Minas Gerais
    @Rule(Fact(trop_cerr_mais7mhab_municipios_maisde500='S'))
    def is_minasgerais(self):
          print('O estado é Minas Gerais.')
    
    #Goiás
    @Rule(Fact(trop_cerr_mais7mhab_municipios_maisde500='N'))
    def is_goias(self):
          print('O estado é o Goiás.')
    
    #pergunta se o estado tem mais de 10 milhoes de habitantes
    @Rule(AND(Fact(clima='tropical'),
              Fact(vegetacao='caatinga')),
              NOT(Fact(trop_caat_habitantes_maisde10m=W())))
    def ask_hab_tropical_caatinga(self):
        self.declare(Fact(trop_caat_habitantes_maisde10m=input("A população do estado tem mais de 10 milhões de habitantes (S/N)?")))
    
    #Bahia
    @Rule(Fact(trop_caat_habitantes_maisde10m='S'))
    def is_bahia(self):
          print('O estado é a Bahia.')
    
    #Sergipe
    @Rule(Fact(trop_caat_habitantes_maisde10m='N'))
    def is_sergipe(self):
          print('O estado é o Sergipe.')
    
    #pergunta se o estado tem mais de 9 milhoes de habitantes
    @Rule(AND(Fact(clima='semiárido'),
              Fact(vegetacao='caatinga')),
              NOT(Fact(sema_caat_habitantes_maisde9m=W())))
    def ask_hab_semiarido_caatinga(self):
        self.declare(Fact(sema_caat_habitantes_maisde9m=input("A população do estado tem mais de 9 milhões de habitantes (S/N)?")))
    
    
    #pergunta se a area do estado tem menos de 100 mil km²
    @Rule(Fact(sema_caat_habitantes_maisde9m='S'),
          NOT(Fact(sema_caat_mais9mhab_area_menos100km=W())))
    def ask_ter_semiarido_caatinga_mais9mhab(self):
          self.declare(Fact(sema_caat_mais9mhab_area_menos100km=input("A área do estado tem menos de 100 mil km²? (S/N)?")))
    
    #Pernambuco
    @Rule(Fact(sema_caat_mais9mhab_area_menos100km='S'))
    def is_pernambuco(self):
          print('O estado é o Pernambuco.')
    
    #Ceará
    @Rule(Fact(sema_caat_mais9mhab_area_menos100km='N'))
    def is_ceara(self):
          print('O estado é o Ceará.')
            
    #pergunta se a area do estado tem mais de 50 mil km²
    @Rule(Fact(sema_caat_habitantes_maisde9m='N'),
          NOT(Fact(sema_caat_menos9mhab_area_mais50km=W())))
    def ask_ter_semiarido_caatinga_menos9mhab(self):
          self.declare(Fact(sema_caat_menos9mhab_area_mais50km=input("A área do estado tem mais de 50 mil km²? (S/N)?")))
    
    #Alagoas
    @Rule(Fact(sema_caat_menos9mhab_area_mais50km='N'))
    def is_alagoas(self):
          print('O estado é o Alagoas.')
            
    #pergunta se o estado tem mais de 200 municipios
    @Rule(Fact(sema_caat_menos9mhab_area_mais50km='S'),
          NOT(Fact(sema_caat_menos9mhab_mais50km_municipios_mais200=W())))
    def ask_ter_semiarido_caatinga_menos9mhab_mais50km(self):
          self.declare(Fact(sema_caat_menos9mhab_mais50km_municipios_mais200=input("O estado tem mais de 200 municipios? (S/N)?")))
    
    
    #Paraíba
    @Rule(Fact(sema_caat_menos9mhab_mais50km_municipios_mais200='S'))
    def is_paraiba(self):
          print('O estado é a Paraíba.')
         
    #Rio Grande do Norte
    @Rule(Fact(sema_caat_menos9mhab_mais50km_municipios_mais200='N'))
    def is_riograndedonorte(self):
          print('O estado é o Rio Grande do Norte.')
            
```

```python
#main
x = 'S'

engine = Estados()
engine.reset()
engine.declare(Fact(action='start'))
engine.run()

while True:
    engine.reset()
    engine.declare(Fact(action='restart'))
    engine.run()
    
while x == 'S':
    engine.reset()
    engine.declare(Fact(ask='S'))
    engine.run()
    x = input(Deseja identificar outro Estado (S/N)?)
    
```
